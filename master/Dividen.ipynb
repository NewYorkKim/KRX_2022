```python
import pandas as pd

df = pd.read_excel('배당정보_유가증권.xlsx', header=2)
df['Unnamed: 1'] = df['Unnamed: 1'].astype(str)
df_2022 = df.where(df['Unnamed: 2'] == 2021).dropna()
df_2022 = df_2022.where(df_2022['주당배당금'] == '-').dropna()
df_no = list(df_2022['Unnamed: 1'].astype(str))
df_tmp = df.where(df['Unnamed: 2'] == 2021).dropna()
b = list(df_tmp['Unnamed: 1'])
df_yes = []
for i in b:
    if i not in df_no:
        df_yes.append(i)

from pykrx import stock

result_yes = stock.get_market_fundamental("20220701", "20220815", "005930", freq="d")['PBR'] * 0
for ticker in df_yes:
    print(ticker)
    try:
        tmp = stock.get_market_fundamental("20220701", "20220815", "005930", freq="d")['PBR'] * 0
        tmp += stock.get_market_fundamental("20220701", "20220815", f'{ticker:0>6}', freq="d")['PBR']
        result_yes += tmp.fillna(0)
        print(result_yes)
    except:
        pass
result_yes = result_yes/len(df_yes)
result2 = stock.get_market_fundamental("20220701", "20220815", "005930", freq="d")['PBR'] * 0
for ticker in df_no:
    print(ticker)
    try:
        tmp = stock.get_market_fundamental("20220701", "20220815", "005930", freq="d")['PBR'] * 0
        tmp += stock.get_market_fundamental("20220701", "20220815", f'{ticker:0>6}', freq="d")['PBR']
        result2 += tmp.fillna(0)
        print(result2)
    except:
        pass
result2 = result2/len(df_no)
```

```python
df2_2022 = pd.read_excel('코스닥_2022.xlsx', header=2)
df2_2022['Unnamed: 1'] = df2_2022['Unnamed: 1'].astype(str)
df_2022_tmp = df2_2022.where(df2_2022['주당배당금'] == '-').dropna()
df2_2022_no = list(df_2022_tmp['Unnamed: 1'])
tmp = list(df2_2022['Unnamed: 1'])
df2_2022_yes = []
for i in tmp:
    if i not in df2_2022_no:
        df2_2022_yes.append(i)
        
from pykrx import stock

result_2022_yes_dac = stock.get_market_fundamental("20220701", "20220815", "005930", freq="d")['PBR'] * 0
for ticker in df2_2022_yes:
    print(ticker)
    try:
        tmp = stock.get_market_fundamental("20220701", "20220815", "005930", freq="d")['PBR'] * 0
        tmp += stock.get_market_fundamental("20220701", "20220815", f'{ticker:0>6}', freq="d")['PBR']
        result_2022_yes_dac += tmp.fillna(0)
        print(result_2022_yes_dac)
    except:
        pass
result_2022_yes_dac = result_2022_yes_dac/len(df2_2022_yes)
result_2022_no_dac = stock.get_market_fundamental("20220701", "20220815", "005930", freq="d")['PBR'] * 0
for ticker in df2_2022_no:
    print(ticker)
    try:
        tmp = stock.get_market_fundamental("20220701", "20220815", "005930", freq="d")['PBR'] * 0
        tmp += stock.get_market_fundamental("20220701", "20220815", f'{ticker:0>6}', freq="d")['PBR']
        result_2022_no_dac += tmp.fillna(0)
        print(result_2022_no_dac)
    except:
        pass
result_2022_no_dac = result_2022_no_dac/len(df2_2022_no)
```
