```python
df = target
df['정제된 댓글'] = df['댓글'].str.replace('\[삭제된 게시물의 답글\]',' ')
df['정제된 댓글'] = df['정제된 댓글'].str.replace('답글:',' ')
df['정제된 댓글'] = df['정제된 댓글'].str.replace('[^가-힣]',' ').str.replace(' +',' ').str.strip()
df = df[df['정제된 댓글'] != '']
df = df.reset_index(drop=True)
print('데이터 크기: ',df.shape)
```

# konlpy

```python
from konlpy.tag import *

hannanum = Hannanum()
kkma = Kkma()
komoran = Komoran()
#mecab = Mecab()
okt = Okt()
```

```python
token_nouns = [hannanum,kkma,komoran,okt]
```

```python
token_set = []
for tokenizing_tool in token_nouns:
    token_set.append([tokenizing_tool.nouns(i) for i in target['정제된 댓글']])
```

```python
token_set[0]
```

```python
nouns_each_method = []
for j in range(len(token_nouns)):
    nouns = []
    for sents_token in token_set[j]:
        nouns.extend(sents_token)
    nouns_each_method.append(nouns)
```

```python
nouns_each_method = [list(set(i)) for i in nouns_each_method]
for i in range(len(nouns_each_method)):
    nouns_each_method[i].sort()
```

```python
tp = list(set(nouns_each_method[0]))
tp.sort()
tp
```

```python
import pandas as pd
table = pd.DataFrame(nouns_each_method).transpose()
table.columns = ['hannanum','kkma','komoran','okt']
table
```

## okt tokenizing

```python
okt = Okt()
tag_list = ['Noun','Verb','Adjective']

tokenized_data = []

for i in range(target.shape[0]):
    tokenized_sentence = okt.pos(target['정제된 댓글'][i], stem=True) # 토큰화
    tag_checked_sentence = []
    for j in tokenized_sentence:
        x,y = j
        if y in tag_list:
            tag_checked_sentence.append(x)
    tokenized_data.append(tag_checked_sentence)
    print(f'\r{i+1}개 형태소분석 완료.',end='')
target['토큰화 댓글'] = tokenized_data
# df = df[df['토큰화 댓글'].str.len() > 1]
# df = df.reset_index(drop=True)
# df = df[df['토큰화 댓글'] != '[]']
# print('\n데이터 크기: ',df.shape)
# target = target.drop('Unnamed: 0',axis=1)
target.head()
```
