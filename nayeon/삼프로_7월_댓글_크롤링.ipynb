```python
api_key = 'api_key'
youtube = build('youtube', 'v3', developerKey=api_key)
channel_id = 'channel_id'
```

```python
videos = scrapetube.get_channel(channel_id)

video_ids = []

for video in videos:
    video_ids.append(video['videoId'])
```

```python
len(video_ids)
```

```python
video_infos = []

for i in range(30):
    start = i * 50 
    end = (i + 1) * 50
    video_request = youtube.videos().list(
        part="snippet",
        id=','.join(video_ids[start:end]))

    video_response = video_request.execute()

    for item in video_response['items']:
        title = item['snippet']['title']
        if ('글로벌 이슈체크' in title) or ('글로벌 마켓브리핑' in title) or ('직장인 vlog' in title):
            continue
        video_infos.append([item['snippet']['title'], item['snippet']['publishedAt'], item['id']])
        
df = pd.DataFrame(video_infos, columns=['title', 'date', 'id'])
df.head()
```

```python
comments_08 = []

for video_id in df_08['id']:
    api_obj = build('youtube', 'v3', developerKey=api_key)
    response = api_obj.commentThreads().list(part='snippet', videoId=video_id, maxResults=100).execute()
 
    while response:
        for item in response['items']:
            comment = item['snippet']['topLevelComment']['snippet']
            comments_08.append([video_id, comment['textDisplay'], comment['authorDisplayName'], comment['publishedAt'], comment['likeCount']])

        if 'nextPageToken' in response:
            response = api_obj.commentThreads().list(part='snippet', videoId=video_id, pageToken=response['nextPageToken'], maxResults=100).execute()
        else:
            break

df0 = pd.DataFrame(comments_08, columns=['id', 'comment', 'author', 'comment_date', 'num_likes'])
print(len(df0))
df0.head()
```

```python
comments_07 = []

for video_id in df_07['id']:
    api_obj = build('youtube', 'v3', developerKey=api_key)
    response = api_obj.commentThreads().list(part='snippet', videoId=video_id, maxResults=100).execute()
 
    while response:
        for item in response['items']:
            comment = item['snippet']['topLevelComment']['snippet']
            comments_07.append([video_id, comment['textDisplay'], comment['authorDisplayName'], comment['publishedAt'], comment['likeCount']])
 
        if 'nextPageToken' in response:
            response = api_obj.commentThreads().list(part='snippet', videoId=video_id, pageToken=response['nextPageToken'], maxResults=100).execute()
        else:
            break

df1 = pd.DataFrame(comments_07, columns=['id', 'comment', 'author', 'comment_date', 'num_likes'])
print(len(df1))
df1.head()
```

```python
sam_recent['date_diff'] = 0

for idx in sam_recent.index:
    v = sam_recent.loc[idx, 'video_date']
    c = sam_recent.loc[idx, 'comment_date']
    diff = c - v
    sam_recent.loc[idx, 'date_diff'] = diff

print(sam_recent.info())
sam_recent.head()
```

```python
import datetime

# 장 시간 
sam_recent['market_hour'] = False
days = ['월요일', '화요일', '수요일', '목요일', '금요일', '토요일', '일요일']

for idx in sam_recent.index:
    h = sam_recent.iloc[idx, 5].hour
    m = sam_recent.iloc[idx, 5].minute
    w = days[sam_recent.iloc[idx, 5].weekday()]
    
    if w != '토요일' and w != '일요일':
        if (datetime.time(h, m) > datetime.time(9, 0)) and (datetime.time(h, m) < datetime.time(15, 30)):
            sam_recent.loc[idx, 'market_hour'] = True
```
