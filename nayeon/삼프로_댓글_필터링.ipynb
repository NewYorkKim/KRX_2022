```python
import pandas as pd
import numpy as np
import datetime
```

```python
sampro_comments = pd.read_csv('./sampro/sampro_comments.csv')
sampro_comments.head()
```

```python
sampro_comments.info()
```

```python
sampro_comments['video_date'].values
```

```python
sampro_comments['comment_date'].values
```

```python
result = []

for idx in sampro_comments.index:
    v = str(sampro_comments.loc[idx, 'video_date']).split()[0]
    c = str(sampro_comments.loc[idx, 'comment_date']).split()[0]
    
    if v == c:
        result.append(sampro_comments.loc[idx, :])
        
len(result)
```

```python
sampro_filtered = pd.DataFrame(result, columns=sampro_comments.columns)
sampro_filtered.head()
```

```python
sampro_filtered.info()
```

```python
sampro_filtered.to_csv('./sampro/sampro_filtered.csv', index=False)
```
