```python
def get_clean_text(df):
    for idx in df.index:
        if(str(df.loc[idx, '댓글']) == 'nan') :
            continue
        else:
            text = df.loc[idx, '댓글']
            clean_text = re.sub(r"[^A-Za-z一-龥가-힣]", " ", text)
            
        df.loc[idx, '댓글'] = clean_text
        
    return df
```

```python
ktop30_dict = ktop30[['종목명', '비중(%)']].set_index('종목명').T.to_dict('records')[0]
ktop30_dict
```

```python
def get_train_data(df, name):
    n = int(30000 * float(ktop30_dict[name]))
    result0 = df.sample(n=n)
    
    result1 = get_clean_text(result0)
    
    result1.to_csv(name+'_train.csv', index=False, encoding='euc-kr')
    
    print(name, '완료')
```
