```python
def get_clean_text(df):
    for idx in df.index:
        if(str(df.loc[idx, 'comment']) == 'nan') :
            continue
        else:
            text = df.loc[idx, 'comment']
            clean_text = re.sub(r"[^A-Za-z一-龥가-힣]", " ", str(text))
            
        df.loc[idx, 'comment'] = clean_text
        
    return df
```

```python
def get_train_data(df):
    n = 10000
    result0 = df.sample(n=n)
    
    result1 = get_clean_text(result0)
    
    result1.to_csv('sampro_train.csv', index=False)
    
    print('완료')
    
    return result1.head()
```
