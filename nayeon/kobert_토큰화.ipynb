```python
from kobert_tokenizer import KoBERTTokenizer
tokenizer = KoBERTTokenizer.from_pretrained('skt/kobert-base-v1')
```

```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/KRX/sampro_train.csv')
df.head()
```

```python
sampro_comments = df[['comment']]
sampro_comments.head()
```

```python
sampro_comments['kobert'] = ''
sampro_comments.head()
```

```python
text = sampro_comments['comment'][0]
tokenizer.tokenize(text)
```

```python
for idx in sampro_comments.index:
  text = sampro_comments['comment'][idx]
  sampro_comments['kobert'][idx] = tokenizer.tokenize(str(text))

sampro_comments
```
