### 만들어진 말뭉치를 이용해 각 댓글에서 명사 추출

```python
comments = []
for c in df["comment"]:
    comments.append(pp.get_nouns(scores,c))
```

```python
for i in range(20):
    print(comments[i])
```

### 단어사전 생성
+ 각 단어별 등장 횟수를 딕셔너리로 생성

```python
word_dict = dict()
for comment in comments:
    for word in comment:
        if word in word_dict:
            word_dict[word] += 1
        else:
            word_dict[word] = 1
```

```python
sorted(word_dict.items(), key = lambda x:x[1], reverse= True)
```
