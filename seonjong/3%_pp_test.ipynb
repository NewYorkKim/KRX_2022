### clean_text 함수 확인

```python
import PreProcesser

pp = PreProcesser.preprocesser()
```

```python
# 특수문자 제거가 정상적으로 작동하는지 확인
text = df["댓글"][1]
print(text)
pp.clean_text(text)

# >> 줄 바꿈 표식인 <br/>에서 br이 제대로 제거되지 않음을 확인
```

```python
text = df["댓글"][3]
print(text)
pp.clean_text(text)

# >> 그 이외의 특수문자 제거 확인
```

```python
# 모든 댓글을 하나의 데이터 프레임에 담기
comment_list = []
for comment in df["댓글"]:
    try:
        tmp_txt = pp.clean_text(comment)
        tmp_txt = re.sub(r'[ㄱ-ㅎㅏ-ㅣ0-9]+','',tmp_txt)
        emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags=re.UNICODE)
        clean_txt = emoji_pattern.sub(r'', tmp_txt)
        comment_list.append(clean_txt)
    except:
        comment_list.append('NaN')

# rsc는 remove special character의 약자
df_remove_sc = pd.DataFrame(comment_list,columns=["comment_rsc"])
```

```python
df_remove_sc.head(20)
```

```python
# 특수문자가 잘 제거되었음을 확인
df_remove_sc = df_remove_sc.dropna()
```

### make_corpus 함수 확인
+ 이 함수는 말뭉치를 생성한 이후 각 단어의 등장 확률을 점수로 저장한 후 반환한다.
+ file path를 직접 입력해서 사용한다.

```python
scores = pp.make_corpus("카카오_2year.csv")
```

### 만들어진 말뭉치를 이용해 각 댓글에서 명사 추출

```python
comments = []
for c in df_remove_sc['comment_rsc']:
    comments.append(pp.get_nouns(scores,c))
```

```python
for i in range(20):
    print(comments[i])
```

### 단어사전 생성
+ 각 단어별 등장 횟수를 딕셔너리로 생성

```python
word_dict = dict()
for comment in comments:
    for word in comment:
        if word in word_dict:
            word_dict[word] += 1
        else:
            word_dict[word] = 1
```
